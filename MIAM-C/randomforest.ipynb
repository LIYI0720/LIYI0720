```python

x=data_copy.iloc[:,2:]
y=data_copy.iloc[:,1]
x.shape,y.shape
```

数据归一化

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features = scaler.fit_transform(x)
```

按照4比1划分训练和验证


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(features, y, test_size=0.2, stratify=y, random_state=0)
```

随机森林模型

```python
#随机森林
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

进行五折交叉验证

```python
from sklearn.model_selection import StratifiedKFold, StratifiedShuffleSplit
import joblib
import os

skf = StratifiedKFold(n_splits=5, shuffle=True,random_state = 0)

rdn_index = skf.split(np.zeros(len(y)),y)


cv_index = rdn_index

for k_fold,(train_index, val_index) in enumerate(rdn_index) :
    train_fea = features[train_index]
    train_target = y.values[train_index]

    val_fea = features[val_index]
    val_target = y.values[val_index]



    np_index = np.array(val_index).reshape(-1,1)
    np_fea = np.array(val_fea)
    np_target = np.array(val_target).reshape(-1,1)
    

    np_val = np.concatenate((np_index,np_fea,np_target),axis=1)
    pd_val = pd.DataFrame(np_val)
    pd_val.to_csv(os.path.join('Z:/admin/LXH/lxh_train_code/machine_learning_lxh_copy/randomFroest_vallist_bc2023', 'cell_val_fea_target' + str(k_fold) + '.csv'))

    model_param = RandomForestClassifier(
                                oob_score=True)
    
    model_param.fit(train_fea,train_target)
    
    joblib.dump(model_param, os.path.join('Z:/admin/LXH/lxh_train_code/machine_learning_lxh_copy/randomFroest_param_bc2023', 'cell_RF_' + str(k_fold) + '.model'))

 
    print("accuracy: %f" % model_param.oob_score_)



```
