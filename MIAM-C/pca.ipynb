```python
# 查看每个主成分的解释方差比例
explained_variance_ratio = pca.explained_variance_ratio_
print(explained_variance_ratio)
```

```python
# 输出结果
for i, evr in enumerate(explained_variance_ratio):
    print(f"Principal Component {i + 1}: Explained Variance Ratio = {evr:.4f}")
```

```python

components = pca.components_

# 查看每个主成分的解释方差比例
explained_variance_ratio = pca.explained_variance_ratio_

# 查看累积的解释方差比例
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# 输出结果
for i, evr in enumerate(explained_variance_ratio):
    print(f"Principal Component {i + 1}: Explained Variance Ratio = {evr:.4f}")

for i, cvr in enumerate(cumulative_variance_ratio):
    print(f"First {i + 1} Principal Components: Cumulative Variance Ratio = {cvr:.4f}")
```
